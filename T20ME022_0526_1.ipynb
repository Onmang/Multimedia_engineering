```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('/content/drive/MyDrive/harcas.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
grayimg = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(grayimg, cmap = 'gray')
```

```python
#2値化
binimg = (grayimg > 180) * 255
plt.imshow(binimg, cmap = 'gray')
```

```python
#シャープ化
filter = np.array([[-1, -1, -1],
                   [-1, 9, -1],
                   [-1, -1, -1]], np.float32)
filtered = cv2.filter2D(grayimg, -1, filter)
plt.imshow(filtered, cmap = 'gray')
```

```python
#ぼかし
filter = np.array([[1/9, 1/9, 1/9],
                   [1/9, 1/9, 1/9],
                   [1/9, 1/9, 1/9]], np.float32)
filtered = cv2.filter2D(grayimg, -1, filter)
plt.imshow(filtered, cmap = 'gray')
```

```python
#エッジ

#横方向
filter = np.array([[-1, 0, 1],
                   [-1, 0, 1],
                   [-1, 0, 1]], np.float32)

#縦方向
filter2 = np.array([[-1, 0, -1],
                   [0, 0, 0],
                   [1, 1, 1]], np.float32)
filtered = cv2.filter2D(grayimg, -1, filter)
filtered2 = cv2.filter2D(grayimg, -1, filter2)

filtered3 = filtered + filtered2
plt.imshow(filtered, cmap = 'gray')
plt.show()
plt.imshow(filtered2, cmap = 'gray')
plt.show()
#横縦組み合わせ
plt.imshow(filtered3, cmap = 'gray')
plt.show()
```
